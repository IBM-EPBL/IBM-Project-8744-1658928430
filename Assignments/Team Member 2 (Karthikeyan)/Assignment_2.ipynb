# 6. Find the outliers and replace the outliers

````python
#identifying the outliers
df.boxplot(column=['Age'], grid=False, color='black')
````

````python
import numpy as np
print(np.where(df['Age']>60))
````

````python
df.boxplot(column=['Age'], grid=False, color='black')
````

````python
df["Age"] = np.where(df["Age"] >60, df['Age'].median(),df['Age'])
df.boxplot(column=['Age'], grid=False, color='black')
````
